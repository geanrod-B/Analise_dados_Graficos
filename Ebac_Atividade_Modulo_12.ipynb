```python
import plotly.express as px

   # --------------------------- GRAFICO 4 BOXPLOT, O GRÁFICO ESCOLHIDO CONFORTE SOLICITADO. -------------------------------



# ------- Dados fornecidos. -------
# Definindo o DataFrame 'df' para garantir que esteja disponível ao executar esta célula.
data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}
df = pd.DataFrame(data)

# ------- BloxPlot de 'inadimplência' e 'Limite de Crédito'. -------

# ------- Convertendo colunas numéricas. -------
df['Limite_Credito'] = df['Limite_Credito'].astype(int)
df['Historico_Inadimplencia'] = df['Historico_Inadimplencia'].astype(int)

# -------- Criando o gráfico. --------
fig = px.box(df, x='Historico_Inadimplencia',
                 y='Limite_Credito',color='Historico_Inadimplencia',
                 title='Boxplot Interativo: Inadimplência x Limite de Crédito',
                 labels={'Historico_Inadimplencia': 'Inadimplência (0 = Não, 1 = Sim)','Limite_Credito': 'Limite de Crédito'})
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


   # --------------------------- GRAFICO 5 BARRAS ILUSTRATIVO -------------------------------



# Convertendo colunas numéricas
df['Limite_Credito'] = df['Limite_Credito'].astype(int)

plt.figure(figsize=(10,5))
sns.barplot(
    x='Profissão',
    y='Limite_Credito',
    data=df,
    estimator='mean',
    ci=None,
    palette='viridis'
)

plt.title('Média do Limite de Crédito por Profissão', fontsize=16)
plt.xlabel('Profissão', fontsize=12)
plt.ylabel('Média do Limite de Crédito', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## **Salário é o principal impulsionador Positivo.**


Observamos que clientes com maior salário tendem a receber limites de crédito mais altos.
Isso confirma que a capacidade financeira é um dos critérios mais relevantes na decisão do limite concedido pelas
instituições.

## **Inadimplência é o fator de maior impacto negativo.**

Independentemente do salário, clientes com histórico de inadimplência recebem limites drasticamente mais baixos. Ou seja, o risco associado ao cliente tem peso elevado na avaliação de crédito.

## **Profissão influencia a estabilidade percebida**.**


Profissões como Medicina apresentaram limites mais altos, enquanto categorias como Estudantes receberam limites menores. Isso indica que a análise de crédito considera também a estabilidade e o potencial de renda da atividade profissional
